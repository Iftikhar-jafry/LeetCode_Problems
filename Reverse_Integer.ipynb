Given a signed 32-bit integer x, return x with its digits reversed. If reversing x causes the value to go outside the signed 32-bit integer range [-231, 231 - 1], then return 0.

Assume the environment does not allow you to store 64-bit integers (signed or unsigned).

 

Example 1:

Input: x = 123
Output: 321
Example 2:

Input: x = -123
Output: -321
Example 3:

Input: x = 120
Output: 21
 

Constraints:

-2^31 <= x <= 2^(31 - 1)


```python
class Solution:
    def reverse(self, x):
        num=x
        if x==0 or  -2**31 > x > 2**31 - 1:
            return 0
        if x<0:
            num=num*-1
        total=0
        while num!=0:
            total=total*10+num%10
            num=num//10
        
        if x<0:
            return total*-1
        return total
        


        
```

```python
sol=Solution()
x=1534236469
print(sol.reverse(x))
```
