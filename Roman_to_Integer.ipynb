<h2>Problem #13: Roman to Integer</h2>

<p>Roman numerals are represented by seven different symbols: I, V, X, L, C, D, and M.</p>

<table>
  <thead>
    <tr>
      <th>Symbol</th>
      <th>Value</th>
    </tr>
  </thead>
  <tbody>
    <tr><td>I</td><td>1</td></tr>
    <tr><td>V</td><td>5</td></tr>
    <tr><td>X</td><td>10</td></tr>
    <tr><td>L</td><td>50</td></tr>
    <tr><td>C</td><td>100</td></tr>
    <tr><td>D</td><td>500</td></tr>
    <tr><td>M</td><td>1000</td></tr>
  </tbody>
</table>

<p>For example:</p>
<ul>
  <li>2 is written as <strong>II</strong> (1 + 1)</li>
  <li>12 is written as <strong>XII</strong> (X + II)</li>
  <li>27 is written as <strong>XXVII</strong> (XX + V + II)</li>
</ul>

<p>Roman numerals are usually written largest to smallest from left to right.  
However, some numerals use <strong>subtractive notation</strong>:</p>
<ul>
  <li>I can be placed before V (5) and X (10) to make 4 (IV) and 9 (IX)</li>
  <li>X can be placed before L (50) and C (100) to make 40 (XL) and 90 (XC)</li>
  <li>C can be placed before D (500) and M (1000) to make 400 (CD) and 900 (CM)</li>
</ul>

<p>Given a Roman numeral, convert it to an integer.</p>

<h3>Examples</h3>

<h4>Example 1:</h4>
<p><strong>Input:</strong> s = "III"</p>
<p><strong>Output:</strong> 3</p>
<p><strong>Explanation:</strong> III = 3</p>

<h4>Example 2:</h4>
<p><strong>Input:</strong> s = "LVIII"</p>
<p><strong>Output:</strong> 58</p>
<p><strong>Explanation:</strong> L = 50, V = 5, III = 3</p>

<h4>Example 3:</h4>
<p><strong>Input:</strong> s = "MCMXCIV"</p>
<p><strong>Output:</strong> 1994</p>
<p><strong>Explanation:</strong> M = 1000, CM = 900, XC = 90, IV = 4</p>

<h3>Constraints:</h3>
<ul>
  <li>1 ≤ s.length ≤ 15</li>
  <li>s contains only the characters ('I', 'V', 'X', 'L', 'C', 'D', 'M')</li>
  <li>It is guaranteed that s is a valid Roman numeral in the range [1, 3999]</li>
</ul>
