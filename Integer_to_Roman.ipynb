<h2>Problem #12: Integer to Roman</h2>

<p>Seven different symbols represent Roman numerals with the following values:</p>

<table>
  <thead>
    <tr>
      <th>Symbol</th>
      <th>Value</th>
    </tr>
  </thead>
  <tbody>
    <tr><td>I</td><td>1</td></tr>
    <tr><td>V</td><td>5</td></tr>
    <tr><td>X</td><td>10</td></tr>
    <tr><td>L</td><td>50</td></tr>
    <tr><td>C</td><td>100</td></tr>
    <tr><td>D</td><td>500</td></tr>
    <tr><td>M</td><td>1000</td></tr>
  </tbody>
</table>

<h3>Rules for Converting Decimal to Roman Numerals</h3>
<ol>
  <li>If the value does not start with 4 or 9:
    <ul>
      <li>Select the symbol of the maximal value that can be subtracted from the input.</li>
      <li>Append that symbol to the result, subtract its value, and convert the remainder recursively.</li>
    </ul>
  </li>
  <li>If the value starts with 4 or 9:
    <ul>
      <li>Use the subtractive form representing one symbol subtracted from the following symbol.</li>
      <li>Examples:
        <ul>
          <li>4 → IV (1 less than 5)</li>
          <li>9 → IX (1 less than 10)</li>
          <li>40 → XL, 90 → XC, 400 → CD, 900 → CM</li>
        </ul>
      </li>
    </ul>
  </li>
  <li>Only powers of 10 (I, X, C, M) can be appended consecutively at most 3 times.</li>
  <li>Symbols like V, L, D cannot be repeated consecutively.</li>
  <li>If you need to append a symbol 4 times, use the subtractive form instead.</li>
</ol>

<h3>Examples</h3>

<h4>Example 1:</h4>
<p><strong>Input:</strong> num = 3749</p>
<p><strong>Output:</strong> "MMMDCCXLIX"</p>
<p><strong>Explanation:</strong></p>
<ul>
  <li>3000 = MMM (1000 + 1000 + 1000)</li>
  <li>700 = DCC (500 + 100 + 100)</li>
  <li>40 = XL (10 less than 50)</li>
  <li>9 = IX (1 less than 10)</li>
</ul>

<h4>Example 2:</h4>
<p><strong>Input:</strong> num = 58</p>
<p><strong>Output:</strong> "LVIII"</p>
<p><strong>Explanation:</strong></p>
<ul>
  <li>50 = L</li>
  <li>8 = VIII (5 + 1 + 1 + 1)</li>
</ul>

<h4>Example 3:</h4>
<p><strong>Input:</strong> num = 1994</p>
<p><strong>Output:</strong> "MCMXCIV"</p>
<p><strong>Explanation:</strong></p>
<ul>
  <li>1000 = M</li>
  <li>900 = CM</li>
  <li>90 = XC</li>
  <li>4 = IV</li>
</ul>

</div>


```python
class Solution:
    def intToRoman(self, num: int) -> str:
        roman_map = [
            (1000, "M"), (900, "CM"), (500, "D"), (400, "CD"),
            (100, "C"), (90, "XC"), (50, "L"), (40, "XL"),
            (10, "X"), (9, "IX"), (5, "V"), (4, "IV"), (1, "I")
        ]
        
        result = ""
        
        for value, symbol in roman_map:
            count = num // value
            if count:
                result += symbol * count
                num -= value * count
        
        return result
```
