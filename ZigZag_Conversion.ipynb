<h2>Problem #6: Zigzag Conversion</h2>

<p>The string "<strong>PAYPALISHIRING</strong>" is written in a zigzag pattern on a given number of rows like this (displayed for clarity):</p>

<pre>
P   A   H   N
A P L S I I G
Y   I   R
</pre>

<p>Then read line by line: "<strong>PAHNAPLSIIGYIR</strong>".</p>

<p>Write a function that takes a string and converts it into a zigzag pattern given a number of rows:</p>

<pre>
string convert(string s, int numRows);
</pre>

<h3>Examples</h3>

<h4>Example 1:</h4>
<p><strong>Input:</strong> s = "PAYPALISHIRING", numRows = 3</p>
<p><strong>Output:</strong> "PAHNAPLSIIGYIR"</p>

<h4>Example 2:</h4>
<p><strong>Input:</strong> s = "PAYPALISHIRING", numRows = 4</p>
<p><strong>Output:</strong> "PINALSIGYAHRPI"</p>
<p><strong>Explanation:</strong></p>
<pre>
P     I    N
A   L S  I G
Y A   H R
P     I
</pre>

<h4>Example 3:</h4>
<p><strong>Input:</strong> s = "A", numRows = 1</p>
<p><strong>Output:</strong> "A"</p>

<h3>Constraints:</h3>
<ul>
  <li>1 ≤ s.length ≤ 1000</li>
  <li>s consists of English letters (lower-case and upper-case), ',' and '.'</li>
  <li>1 ≤ numRows ≤ 1000</li>
</ul>

</div>


```python
class Solution:
    def convert(s, numRows):
        sign=+1
        ind=0
        row=0
        col=0
        res=""
        arr=[[] for _ in range(numRows)]
        print(len(arr[0]))
        while ind< len(s):
            arr[row].append(s[ind])
            print(row)
            row=row+sign
            if row>numRows-1:
                row=row-2
                sign=-1*sign
                col=col+1
            elif row<0:
                row=row+2
                sign=-1*sign
                col=col+1
            ind=ind+1
        
        for i in range(len(arr)):
            for j in range(len(arr[i])):
                res=res+arr[i][j]
        print(res)
```

```python
sol=Solution
sol.convert("123456789", 3)
```
