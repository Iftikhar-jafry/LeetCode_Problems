## Problem#2
#### Add Two Numbers
#### You are given two non-empty linked lists representing two non-negative integers. The digits are stored in reverse order, and each of their nodes contains a single digit. Add the two numbers and return the sum as a linked list.

You may assume the two numbers do not contain any leading zero, except the number 0 itself.
Example 1:
* Input: l1 = [2,4,3], l2 = [5,6,4]
* Output: [7,0,8]
* Explanation: 342 + 465 = 807.

```python
class ListNode:
     def __init__(self, val=0, next=None):
         self.val = val
         self.next = next
class Solution:
    def addTwoNumbers(self, l1, l2):
        curr1=l1
        curr2=l2
        num1=0
        num2=0
        carry=1
        while curr1!= None:
            num1=num1+curr1.val *carry
            curr1=curr1.next
            carry=carry*10
        carry=1
        while curr2!= None:
            num2=num2+curr2.val *carry
            carry=carry*10
            curr2=curr2.next
        total=num1+num2
        head=ListNode()
        curr=head
        if total==0:
            return ListNode()
        while total!=0:
            val=total%10
            total=total//10
            node=ListNode(val)
            curr.next=node
            curr=node
        head=head.next
        while head!=None:
            print(head.val)
            head=head.next
```

```python
# l1[4,5,6]
l1=ListNode(4)
l2=ListNode(5)
l3=ListNode(6)
l1.next=l2
l2.next=l3

# w1[5,3,2]
w1=ListNode(5)
w2=ListNode(3)
w3=ListNode(2)
w1.next=w2
w2.next=w3

sol=Solution()
sol.addTwoNumbers(l1,w1) # 654+235 
```
