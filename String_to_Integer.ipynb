<h2>Problem #8: String to Integer (atoi)</h2>

<p>Implement the <code>myAtoi(string s)</code> function, which converts a string to a 32-bit signed integer.</p>

<h3>Algorithm:</h3>
<ol>
  <li><strong>Whitespace:</strong> Ignore any leading whitespace (" ").</li>
  <li><strong>Signedness:</strong> Determine the sign by checking if the next character is '-' or '+', assuming positivity if neither is present.</li>
  <li><strong>Conversion:</strong> Read the integer by skipping leading zeros until a non-digit character is encountered or the end of the string is reached. If no digits were read, the result is 0.</li>
  <li><strong>Rounding:</strong> If the integer is out of the 32-bit signed integer range <code>[-2³¹, 2³¹ - 1]</code>, clamp it to the range. Integers less than -2³¹ become -2³¹, and integers greater than 2³¹-1 become 2³¹-1.</li>
  <li><strong>Return:</strong> Return the integer as the final result.</li>
</ol>

<h3>Examples</h3>

<h4>Example 1:</h4>
<p><strong>Input:</strong> s = "42"</p>
<p><strong>Output:</strong> 42</p>
<p><strong>Explanation:</strong></p>
<ul>
  <li>Step 1: "42" (no leading whitespace)</li>
  <li>Step 2: "42" (no '-' or '+', assume positive)</li>
  <li>Step 3: "42" ("42" is read in)</li>
</ul>

<h4>Example 2:</h4>
<p><strong>Input:</strong> s = " -042"</p>
<p><strong>Output:</strong> -42</p>
<p><strong>Explanation:</strong></p>
<ul>
  <li>Step 1: Leading whitespace ignored</li>
  <li>Step 2: '-' is read, result should be negative</li>
  <li>Step 3: "042" is read (leading zeros ignored)</li>
</ul>

<h4>Example 3:</h4>
<p><strong>Input:</strong> s = "1337c0d3"</p>
<p><strong>Output:</strong> 1337</p>
<p><strong>Explanation:</strong> Reading stops at the first non-digit character after "1337".</p>

<h4>Example 4:</h4>
<p><strong>Input:</strong> s = "0-1"</p>
<p><strong>Output:</strong> 0</p>
<p><strong>Explanation:</strong> Only "0" is read; reading stops at '-'.</p>

<h4>Example 5:</h4>
<p><strong>Input:</strong> s = "words and 987"</p>
<p><strong>Output:</strong> 0</p>
<p><strong>Explanation:</strong> Reading stops at the first non-digit character 'w'.</p>

<h3>Constraints:</h3>
<ul>
  <li>0 ≤ s.length ≤ 200</li>
  <li>s consists of English letters (lower-case and upper-case), digits (0-9), ' ', '+', '-', and '.'</li>
</ul>

</div>


```python
class Solution:
    def myAtoi(self, s: str) -> int:
        s=s.lstrip()
        if not s:
            return 0
        sign =1
        ind=0
        if s[ind]=="-":
            sign=-1
            ind+=1
        elif s[ind]=="+":
            ind+=1
        elif not s[ind].isdigit():
            return 0
        result=0
        while ind < len(s) and s[ind].isdigit():
            result=result*10+int(s[ind])
            ind+=1
        result=result*sign
        if result< -2**31:
            return -2**31
        elif result > 2**31-1:
            return 2**31-1
        return result
```

```python
sol=Solution()
print(sol.myAtoi("324"))
print(sol.myAtoi("+324"))
print(sol.myAtoi("-324"))
print(sol.myAtoi(" -324"))
print(sol.myAtoi("   w324"))
print(sol.myAtoi("0324"))
print(sol.myAtoi("32e4"))
```
