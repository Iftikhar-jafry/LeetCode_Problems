<h2>Problem #2: Add Two Numbers</h2>

<p>You are given two non-empty linked lists representing two non-negative integers. The digits are stored in <strong>reverse order</strong>, and each of their nodes contains a single digit. Add the two numbers and return the sum as a linked list.</p>

<p>You may assume the two numbers do not contain any leading zero, except the number 0 itself.</p>

<h3>Examples</h3>

<h4>Example 1:</h4>
<p><strong>Input:</strong> l1 = [2,4,3], l2 = [5,6,4]</p>
<p><strong>Output:</strong> [7,0,8]</p>
<p><strong>Explanation:</strong></p>
<ul>
  <li>The numbers are 342 (from [2,4,3]) and 465 (from [5,6,4]).</li>
  <li>Their sum is 807, which is represented in reverse order as [7,0,8].</li>
</ul>

<h4>Example 2:</h4>
<p><strong>Input:</strong> l1 = [0], l2 = [0]</p>
<p><strong>Output:</strong> [0]</p>
<p><strong>Explanation:</strong> Both numbers are 0, sum is 0.</p>

<h4>Example 3:</h4>
<p><strong>Input:</strong> l1 = [9,9,9,9,9,9,9], l2 = [9,9,9,9]</p>
<p><strong>Output:</strong> [8,9,9,9,0,0,0,1]</p>
<p><strong>Explanation:</strong></p>
<ul>
  <li>The numbers are 9,999,999 (from l1) and 9,999 (from l2).</li>
  <li>Their sum is 10,009,998, which is represented in reverse order as [8,9,9,9,0,0,0,1].</li>
</ul>
